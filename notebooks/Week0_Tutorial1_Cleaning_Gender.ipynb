```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('EmergencyTriageDataset_Reduced_Dirty.csv')
df.head()
```

```python
df['Gender'].unique()
df['Gender'].value_counts()
```

```python
gender_map = {
    'Male': 1,
    'MALE': 1,
    '1': 1,
    'Female': 0,
    'FEMALE': 0,
    '0': 0
}
```

```python
df['Gender_Clean'] = df['Gender'].map(gender_map)
```

```python
df['Gender_Clean'].value_counts()
df['Gender_Clean'].isnull().sum()
```

```python
df = df.drop(columns=['Gender'])
df = df.rename(columns={'Gender_Clean': 'Gender'})
```

```python
df.head()
df['Gender'].value_counts()
```

```python
df.to_csv('EmergencyTriageDataset_Reduced_Clean.csv', index=False)
```

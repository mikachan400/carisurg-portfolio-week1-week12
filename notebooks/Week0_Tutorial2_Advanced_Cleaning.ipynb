```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
# Week 0 - Day 2: Advanced Data Cleaning

# In this notebook, multiple clinical columns including GCS, SBP, Temperature,
# and pulse is cleaned using standard data cleaning techniques from a triage dataset.

print("Python version:", sys.version)
```

```python
# Mount your Google Drive so files are accessible from Colab

from google.colab import drive
drive.mount('/content/drive')
print("Drive mounted successfully!")
FILE_PATH = '/content/drive/MyDrive/ColabNotebooks/EmergencyTriageDataset_Reduced_Dirty.csv'

#FILE_PATH = 'EmergencyTriageDataset_Reduced_Dirty.csv'

df = pd.read_csv(FILE_PATH)

# Re-apply Gender cleaning from Day 1 so we start from a consistent state

gender_map = {'Male': 1, 'MALE': 1, '1': 1, 'Female': 0, 'FEMALE': 0, '0': 0}
df['Gender'] = df['Gender'].map(gender_map)

print("Gender cleaned:\n")
print(df['Gender'].value_counts())
print(f"\nDataset: {df.shape[0]} rows x {df.shape[1]} columns")
```

```python
# Inspect before cleaning
# The five(5) useful Inspection commands

print("=== head() — first 5 rows ===")
print(df.head())

print("\n=== tail() — last 5 rows ===")
print(df.tail())

print("\n=== info() — column types ===")
df.info()

print("\n=== describe() — summary stats ===")
print(df.describe())

print("\n=== shape - (rows, columns) ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# Clean GCS
# GCS should be between 3 and 15

print("Unique GCS values:")
print(df['GCS'].unique())

print("\nMissing values before:", df['GCS'].isnull().sum())

# Convert to numeric
# Non-numeric values (e.g. text or errors) are converted to NaN
df['GCS'] = pd.to_numeric(df['GCS'], errors='coerce')

# Remove invalid values
df.loc[df['GCS'] < 3, 'GCS'] = np.nan
df.loc[df['GCS'] > 15, 'GCS'] = np.nan

# Fill missing values with median
df['GCS'] = df['GCS'].fillna(df['GCS'].median())

# Final check
print(df['GCS'].describe())
print("NaNs after:", df['GCS'].isnull().sum())

# After cleaning, all values fall within valid clinical range and no missing values remain.
```

```python
# Clean SBP
# SBP should be between 50 and 250 mmHg

print("Unique SBP values before:")
print(df['SBP'].unique())

print("\nMissing values before:", df['GCS'].isnull().sum())

# Convert to numeric
# Non-numeric values (e.g. text or errors) are converted to NaN
df['SBP'] = pd.to_numeric(df['SBP'], errors='coerce')

# Remove invalid values
df.loc[(df['SBP'] < 50) | (df['SBP'] > 250), 'SBP'] = np.nan

# Fill missing values with median
df['SBP'] = df['SBP'].fillna(df['SBP'].median())

print(df['SBP'].describe())
print("NaNs:", df['SBP'].isnull().sum())

# After cleaning, all values fall within valid clinical range and no missing values remain.
```

```python
# Clean Temperature
# Temperature should be between 32 and 43°C

def to_celsius(val):
    if pd.isnull(val):
        return np.nan
    val = str(val).replace('C', '').strip()
    try:
        return float(val)
    except:
        return np.nan

print("Unique Temp values before:")
print(df['Temp'].unique())

print("\nMissing values before:", df['GCS'].isnull().sum())

df['Temp'] = df['Temp'].apply(to_celsius)

df.loc[(df['Temp'] < 32) | (df['Temp'] > 43), 'Temp'] = np.nan

df['Temp'] = df['Temp'].fillna(df['Temp'].median())

print(df['Temp'].describe())
print("NaNs:", df['Temp'].isnull().sum())

# After cleaning, all values fall within valid clinical range and no missing values remain.
```

```python
# Chosen column
# Clean Pulse
# Pulse should be between 20 and 250 bpm

print("Unique pulse values before:")
print(df['pulse'].unique())

print("\nMissing values before:", df['GCS'].isnull().sum())

# Convert to numeric
# Non-numeric values (e.g. text or errors) are converted to NaN
df['pulse'] = pd.to_numeric(df['pulse'], errors='coerce')

# Remove invalid values
df.loc[(df['pulse'] < 20) | (df['pulse'] > 250), 'pulse'] = np.nan

# Fill missing values with median
df['pulse'] = df['pulse'].fillna(df['pulse'].median())

print(df['pulse'].describe())
print("NaNs:", df['pulse'].isnull().sum())

# After cleaning, all values fall within valid clinical range and no missing values remain.
```
